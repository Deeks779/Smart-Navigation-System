```python
import folium
import pandas as pd
from folium import plugins
import json
import ipywidgets
import os
print("done")
```

```python
#to switch co-ordinate
def switchPosition(coordinate):
        temp = coordinate[0]
        coordinate[0] = coordinate[1]
        coordinate[1] = temp
        return coordinate

#to load path files in a dictionary
def dict_paths(path_add):
    dict_path_files={}
    for root, dirs, files in os.walk(path_add):
            for file in files:
                dict_path_files[file.split('.')[0]] = root+'/'+file
    return dict_path_files

#to change the starting location
def change_start(whereFrom):
    Navigator.changeStartPoint(whereFrom)

#to change the ending location
def displayWay(whereTo):
    Navigator.changeDestination(whereTo)
    #Navigator.drawPathWay(m)
    #Navigator.draw_map(m,list_col)
    #display(m)

def change_start_floor(whereFrom):
    Floor.change_StartPoint(whereFrom)

def display_Way(whereTo):
    Floor.change_Destination(whereTo)

def changeinfo_floor(whereFrom):
    Floor.chnage_info_floor(whereFrom)

def change_start_floor1(whereFrom):
    Floor1.change_StartPoint1(whereFrom)

def display_Way1(whereTo):
    Floor1.change_Destination1(whereTo)

def changeinfo_floor1(whereFrom):
    Floor1.chnage_info_floor1(whereFrom)

def change_start_floor2(whereFrom):
    Floor2.change_StartPoint2(whereFrom)

def display_Way2(whereTo):
    Floor2.change_Destination2(whereTo)

def changeinfo_floor2(whereFrom):
    Floor2.chnage_info_floor2(whereFrom)

print("done")
```

```python
class navigator:
    def __init__(self):
        self.start = 'g'
        self.destination = 'a'
        self.loc=(23.077386057472637,76.85258607068243)

    def changeDestination(self,new_destination):
        self.destination = new_destination
        self.draw_map()
        print(f'Selected Start: {self.start} \nSelected Destination: {self.destination}')

    def changeStartPoint(self, new_startpoint):
        self.start = new_startpoint
        #self.draw_map()

    def drawPathWay(self,m):

        path_add='C:/Users/Deeksha/Desktop/Project ex1/Paths'
        file_dict=dict_paths(path_add)

        search_file = self.start + self.destination
        with open(file_dict[search_file]) as f:
            testWay = json.load(f)

        for feature in testWay['features']:
            path = feature['geometry']['coordinates']

        finalPath = list(map(switchPosition,path))
        folium.plugins.AntPath(finalPath).add_to(m)

    def add_building(self,m):
        list_col={'C:/Users/Deeksha/Desktop/Project ex1/Red':'#FF0000','C:/Users/Deeksha/Desktop/Project ex1/Pink':'#C154C1','C:/Users/Deeksha/Desktop/Project ex1/Blue':'#0000FF','C:/Users/Deeksha/Desktop/Project ex1/Orange':'#FFA500','C:/Users/Deeksha/Desktop/Project ex1/Green':'#008000'}
        for i in list_col:
            dict_paths(i)
            for j in dict_paths(i):
                with open(dict_paths(i)[j]) as f:
                    t = json.load(f)
                folium.Choropleth(geo_data=t,fill_color=list_col[i],line_color='black').add_to(m)

    def add_marker(self,m):
        df=pd.read_csv('C:/Users/Deeksha/Desktop/Project ex1/main_map.csv')
        for i,row in df.iterrows():
            iframe=folium.IFrame(str(row["name"],))
            popup=folium.Popup(iframe,min_width=150,max_width=150)
            folium.Marker(location=[row["lat"],row["lon"]],popup=popup).add_to(m)

    def draw_map(self):
        m=folium.Map(location=self.loc, width="75%", zoom_start= 17)
        folium.TileLayer('stamentoner').add_to(m)
        self.drawPathWay(m)
        self.add_building(m)
        self.add_marker(m)
        display(m)

Navigator = navigator()
print("done")
```
